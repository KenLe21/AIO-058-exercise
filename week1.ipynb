```python
import numpy as np
def lightness_to_gray(img):
    # Tnh gi tr max v min ca cc knh mu
    max_color = np.max(img[..., :3], axis=-1)
    min_color = np.min(img[..., :3], axis=-1)
    # Tnh gi tr xm
    gray_img = (max_color + min_color) / 2
    return gray_img

# Chuyn i nh mu sang nh xm
gray_img_01 = lightness_to_gray(img)

# In gi tr ca pixel (0, 0)
print(gray_img_01[0, 0])
```

```python
# p dng phng php Average  chuyn i nh mu sang nh xm
import matplotlib.pyplot as plt
def average_to_gray(img):
    # Tnh gi tr trung bnh ca cc knh mu
    gray_img = np.mean(img[..., :3], axis=-1)
    return gray_img

# Chuyn i nh mu sang nh xm
gray_img_02 = average_to_gray(img)

# Hin th nh xm
plt.imshow(gray_img_02, cmap='gray')
plt.axis('off')
plt.show()

# In gi tr ca pixel (0, 0)
print(gray_img_01[0, 0])
```

```python
def luminosity_to_gray(img):
    # nh ngha trng s cho cc knh mu
    weights = [0.21, 0.72, 0.07]
    # Tnh gi tr xm da trn trng s
    gray_img = np.dot(img[..., :3], weights)
    return gray_img

gray_img_03 = luminosity_to_gray(img)

# Hin th nh xm
plt.imshow(gray_img_03, cmap='gray')
plt.axis('off')
plt.show()

# In gi tr ca pixel (0, 0)
print(gray_img_03[0, 0])
```

```python
import pandas as pd
df = pd.read_csv("/content/advertising.csv")

data = df . to_numpy ()
pd.DataFrame(data)
```

```python
df_from_numpy = pd.DataFrame(data)


new_column_names = ['TV', 'Radio', 'Newspaper', 'Sales']
df_from_numpy.columns = new_column_names

print(df_from_numpy)
```

```python
# Tm gi tr ln nht v ch mc ca n
max_value = data[:, 3].max()
max_index = data[:, 3].argmax()
print(f"Max: {max_value} - Index: {max_index}")
```

```python
mean_tv = df['TV'].mean()
print(f"Gi tr trung bnh ca ct TV l: {mean_tv}")
```

```python
count_sales = (df['Sales'] >= 20).sum()  # df[df['Sales'] >= 20]['Sales'].count()
print(f"S lng bn ghi c gi tr ti ct Sales ln hn hoc bng 20 l: {count_sales}")

```

```python
# Tnh gi tr trung bnh ca ct Radio khi gi tr ct Sales >= 15
mean_radio_condition = df[df['Sales'] >= 15]['Radio'].mean()
print(f"Gi tr trung bnh ca ct Radio khi gi tr ct Sales >= 15 l: {mean_radio_condition}")
```

```python
# Tnh gi tr trung bnh ca ct Newspaper
mean_newspaper = df['Newspaper'].mean()

# Tnh tng cc hng ca ct Sales vi iu kin Newspaper > gi tr trung bnh ca Newspaper
total_sales_condition = df[df['Newspaper'] > mean_newspaper]['Sales'].sum()
print(f"Tng cc hng ca ct Sales vi iu kin Newspaper > gi tr trung bnh ca Newspaper l: {total_sales_condition}")
```

```python
A = df['Sales'].mean()

# To mng scores da trn gi tr trung bnh
def categorize_sales(value, avg):
    if value > avg:
        return "Good"
    elif value < avg:
        return "Bad"
    else:
        return "Average"

scores = df['Sales'].apply(lambda x: categorize_sales(x, A))

# In ra kt qu ca scores[7:10]
print(scores[7:10].tolist())
```
